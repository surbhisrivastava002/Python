```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv(r'C:\Users\Suresh\Downloads\cardekho.csv')
```

```python
dataset.head(2)
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
dataset.dropna(inplace=True)
```

# name

```python
dataset['car_name']=dataset['name'].apply(lambda x:x[:x.index(' ')])
```

```python
dataset['name']=dataset['name'].apply(lambda x:x[x.index(' ')+1:])
```

```python
fig=plt.figure(figsize=(5,4))
ax=sns.countplot(x='transmission',data=dataset,hue='transmission')
for x in ax.containers:
    ax.bar_label(x)
ax.set_title('Transmittion of car')
plt.show()
```

```python
fig=plt.figure(figsize=(5,4))
ax=sns.countplot(x='fuel',data=dataset,hue='fuel')
for x in ax.containers:
    ax.bar_label(x)
ax.set_title('Different fuel type')
plt.show()
```

```python
ax=sns.countplot(x='seats',data=dataset)
for x in ax.containers:
    ax.bar_label(x)
ax.set_title('Car related to Seats')
```
