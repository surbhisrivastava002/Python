```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel(r'C:\Users\Suresh\Downloads\Apocolypse Food Prep - Power Query Editor Tutorial.xlsx')
```

```python
fig=plt.figure(figsize=(20,7))
data=df.groupby(by=['Store','Product'])['Price'].sum().sort_values(ascending=False).reset_index()
ax=sns.barplot(x='Store',y='Price',data=data,hue='Product')
ax.set_xlabel('Store',fontsize=15)
ax.set_ylabel('Price',fontsize=15)
ax.set_title('No. of Product in each Store vs Price',fontsize=20)
for x in ax.containers:
    ax.bar_label(x)
plt.show()
```
