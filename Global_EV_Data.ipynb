# Value

```python
dataset['value']=round(dataset['value'],2)
```

```python
reg=dataset.groupby(by=['region'])['mode'].count().sort_values(ascending=False).reset_index().head()
ax=sns.barplot(x='region',y='mode',data=reg,hue='region')
for x in ax.containers:
    ax.bar_label(x)
ax.set_title('Region vs Mode')
plt.show()
```

```python
cat=dataset.groupby(by=['category'])['mode'].count().sort_values(ascending=False).reset_index()
ax=sns.barplot(x='category',y='mode',data=cat,hue='category')
for x in ax.containers:
    ax.bar_label(x)
ax.set_title('Category vs Mode')
plt.show()
```

```python
fig=plt.figure(figsize=(25,7))
par=dataset.groupby(by=['parameter'])['mode'].count().reset_index()
ax=sns.barplot(x='parameter',y='mode',data=par,hue='parameter')
for x in ax.containers:
    ax.bar_label(x)
ax.set_title('Parameter vs Mode')
plt.show()
```
